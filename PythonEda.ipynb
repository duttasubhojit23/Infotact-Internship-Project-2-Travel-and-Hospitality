```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('SQL Output Tables.xlsx', sheet_name='Revenue by room type')

# Distribution
sns.histplot(df['revenue'], kde=True)
plt.title('Revenue Distribution')
plt.show()

# Correlation
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_excel('SQL Output Tables.xlsx', sheet_name='Customer Type Analysis')

# Distribution
sns.histplot(df['total_bookings'], kde=True)
plt.title('Total Bookings Distribution')
plt.show()

# Correlation
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_excel('SQL Output Tables.xlsx', sheet_name='Monthly Booking Trend')

# Line Plot (Trend)
sns.lineplot(x='month', y='bookings', data=df)
plt.title('Monthly Booking Trend')
plt.show()

# Distribution
sns.histplot(df['bookings'], kde=True)
plt.title('Bookings Distribution')
plt.show()

# Correlation
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_excel('SQL Output Tables.xlsx', sheet_name='Bookings By City')

# Bar Plot
sns.barplot(x='City', y='total_bookings', data=df)
plt.xticks(rotation=45)
plt.title('Bookings by City')
plt.show()

# Distribution
sns.histplot(df['total_bookings'], kde=True)
plt.title('Bookings Distribution')
plt.show()

# Correlation
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_excel('SQL Output Tables.xlsx', sheet_name='Lead TIme vs Cancellation rate')

# Scatter Plot
sns.scatterplot(x='lead_time', y='cancellation_rate', data=df)
plt.title('Lead Time vs Cancellation Rate')
plt.show()

# Distribution
sns.histplot(df['lead_time'], kde=True)
plt.title('Lead Time Distribution')
plt.show()

sns.histplot(df['cancellation_rate'], kde=True)
plt.title('Cancellation Rate Distribution')
plt.show()

# Correlation
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```
